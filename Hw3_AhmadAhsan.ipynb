### 11b. More complex operations on data for which built-in functions may not exist – using DF.apply()

```python
popDF[['y2019']].apply(lambda x : np.exp((x - x.mean() )/ x.std())).head()
```

```python
# Writing the same command as above using a custom function

def expstandardize(x):
    return np.exp((x - x.mean() )/ x.std())

popDF[['y2019']].apply(expstandardize).head()
```

```python
# Same calculation on multiple columns
popDF[['y2018', 'y2019']].apply(expstandardize).head()
```

### Answer 11.ii

```python
popDF['coeffofvar'] = popDF.loc[: , 'y2010' : 'y2019'].apply(lambda x : np.exp(x.std() / x.mean()), axis = 1)
popDF['coeffofvar'].head(5)
```

### 11c. Aggregation operations using DF.agg()

```python
# we want to calculate a bunch of different statistics for a few numeric columns – y2019 and y2018.
popDF[['y2019', 'y2018']].agg(['count', 'sum', 'min', 'max', 'mean', 'var', 'skew', 'kurt', 'median','idxmax', 'idxmin'])
```

```python
# If we want to perform different aggregation operation for different columns in a DF
popDF[['y2019', 'y2018']].agg({'y2019': ['min', 'max'], 'y2018' : ['var', 'std']})
```

### 11d. Performing grouping operations using DF.groupby()

```python
# we want to compute the mean population for two years (2010 and 2019) within each Region.
popDF[['Region', 'y2010', 'y2019']].groupby(['Region']).mean().sort_values(['Region'])
```

```python
# we want to compute several aggregation statistics for the column y2010 and y2019
popDF.groupby(['Region'])[['y2019', 'y2010']].agg(['min', 'idxmin', 'max', 'idxmax', 'mean'])
```
